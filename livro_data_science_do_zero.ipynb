# Histograma com grafico de barras

```python
from collections import Counter
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
histogram = Counter(min(grade // 10 * 10,90)for grade in grades)
plt.bar([x + 5 for x in histogram.keys()],
        histogram.values(), 
        10, 
        edgecolor=(0, 0, 0))
plt.axis([-5, 105, 0, 5])
plt.xticks([10 * i for i in range(11)])
plt.xlabel('Notas')
plt.ylabel('N° de Alunos')
plt.title(' Distribuição de Notas da 1º Avaliação')
plt.show()
```

```python
import numpy as np
from collections import Counter

# Lista de notas
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]

# Cálculos estatísticos
media = np.mean(grades)  # Média
moda = Counter(grades).most_common(1)[0][0]  # Moda (valor mais frequente)
minimo = np.min(grades)  # Mínimo
mediana = np.median(grades)  # Mediana
total_alunos = len(grades)  # Total de alunos

# Exibir os resultados
print(f'Média: {media}')
print(f'Moda: {moda}')
print(f'Mínimo: {minimo}')
print(f'Mediana: {mediana}')
print(f'Total de Alunos: {total_alunos}')
```

```python
import matplotlib.pyplot as plt
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

plt.plot(xs, variance,  'g-', label='variance')
plt.plot(xs, bias_squared, 'r-', label='bias^2')
plt.plot(xs, total_error, 'b:', label='total erro')

plt.legend(loc=9)
plt.xlabel("Complexidade do Modelo")
plt.xticks([])
plt.title(" O Dilema Viés-Variância")
plt.show()
```

# Compreeendendo o codigo

Definição das listas de dados
Criamos três listas:

variance: Representa a variância do modelo (o quanto ele muda ao tentar se ajustar aos dados).

bias_squared: Representa o erro de viés ao quadrado (o quão longe o modelo está da realidade por simplificações excessivas).

total_error: Soma da variância e do viés ao quadrado para cada valor correspondente.

variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]

**python**

Plotagem dos gráficos
Cada conjunto de dados é representado por uma linha colorida:

Variância em verde ('g-', linha sólida).

Viés ao quadrado em vermelho ('r-', linha sólida).

Erro total em azul ('b:', linha pontilhada).

**python**

Adição da legenda

Personalização do gráfico
O eixo X recebe um rótulo indicando a complexidade do modelo. O comando plt.xticks([]) remove os valores do eixo X, deixando o gráfico mais limpo. Também adicionamos um título para identificar a análise.

# Gráfico de dispersão
O gráfico de dispersão é a opção certa para representar as relações entre pares de conjunto de dados.

```python
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
plt.scatter(friends, minutes)# cria o gráfico de dispersãos
#adiciona o rótulo de cade ponto
for label, friends_count, minutes_count in zip(labels, friends, minutes):
    plt.annotate(label,
    xy=(friends_count, minutes_count),
    xytext=(5, -5),
    textcoords='offset points')
#personalizações do gráfico
plt.title("Minutos Diarios x Numero de Amigos")
plt.xlabel('N° de Amigos')
plt.ylabel(" Minutos Diarios Dedicados ao Site")
plt.show() #exibe o gráfico so uma vez.
```

```python
from typing import List

Vector = List[float]
#def media(valores: vector) -> float:
    #return sum(valores) / len(valores)
heigth_weigth_age=[70, 170, 40]# polegadas, libras, anos
#person_info = {"height_inches": 70, "weight_lbs": 170, "age": 40}
grades = [95, 80, 75, 62]#teste 1,2,3,4
def add(v: Vector, w: Vector) -> Vector:
    assert len(v) == len(w), "Os vetores devem ter o mesmo comprimento."
    return [v_i + w_i for v_i, w_i in zip(v, w)]
assert add([1, 2, 3], [4, 5, 6]) == [5, 7, 9]
import matplotlib.pyplot as plt
plt.plot([1, 2, 3], [4, 5, 6])
plt.show()
```

 As vezes é preciso somar uma lista de vetores por compronentes

```python
def vector_sum(vectors: List[Vector]) -> Vector:
    assert vectors, "no vectors provided"#sem vetor vazio
    #verifica se os vetores são do mesmo tamanho.
    num_elements = len(vectors[0])
    assert all(len(v) == num_elements for v in vectors), "Os vetores devem ter o mesmo comprimento."
    # o elemento de n° do resultado é a soma de todos os vectores[i]
    return [sum(v[i] for v in vectors)
           for i in range(num_elements)]
assert vector_sum([[1, 2], [3, 4], [5, 6 ], [7, 8]]) == [16, 20]


```
