# Gráfico de linha Simples

```python
from matplotlib import pyplot as plt
years = (1950, 1960, 1970, 1980, 1990, 2000, 2010)
gdp = (300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3)
plt.plot(years, gdp,  color='green', marker='o', linestyle= 'solid')
```

## **marker**
Aqui estão alguns exemplos de valores para marker:

'o' → Círculo

's' → Quadrado

'^' → Triângulo para cima

'v' → Triângulo para baixo

'x' → Cruz

'D' → Losango

## **linetyle**
Aqui estão alguns exemplos de valores que você pode usar:

'solid' → Linha contínua (padrão)

'dashed' → Linha tracejada (--)

'dotted' → Linha pontilhada (:)

'dashdot' → Linha com traço e ponto alternados (-.)

# Gráfico de barra simples

```python
movies = ['Annie Hall', 'Ben-Hur', 'Casablanca',
          'Gandhi', 'West Side Story']
num_oscar = [5, 11, 3, 8, 10]
plt.bar(range(len(movies)), num_oscar)
plt.title("My Favority Movies")
plt.ylabel('N° de Oscars')
plt.xticks(range(len(movies)), movies)
plt.show()
```

# Histograma com grafico de barras

```python
from collections import Counter
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
histogram = Counter(min(grade // 10 * 10,90)for grade in grades)
plt.bar([x + 5 for x in histogram.keys()],
        histogram.values(), 
        10, 
        edgecolor=(0, 0, 0))
plt.axis([-5, 105, 0, 5])
plt.xticks([10 * i for i in range(11)])
plt.xlabel('Notas')
plt.ylabel('N° de Alunos')
plt.title(' Distribuição de Notas da 1º Avaliação')
plt.show()
```

```python
import numpy as np
from collections import Counter

# Lista de notas
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]

# Cálculos estatísticos
media = np.mean(grades)  # Média
moda = Counter(grades).most_common(1)[0][0]  # Moda (valor mais frequente)
minimo = np.min(grades)  # Mínimo
mediana = np.median(grades)  # Mediana
total_alunos = len(grades)  # Total de alunos

# Exibir os resultados
print(f'Média: {media}')
print(f'Moda: {moda}')
print(f'Mínimo: {minimo}')
print(f'Mediana: {mediana}')
print(f'Total de Alunos: {total_alunos}')
```
